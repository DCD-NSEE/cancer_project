```python
df_gmm = pd.read_csv('/content/drive/MyDrive/Trabalho/Cancer/Datasets/gmm_tsne.csv')
print(df_gmm.shape)
df_gmm.head(3)
```

```python
df_gmm.grupo_gmm.value_counts()
```

```python
k_gmm = df_gmm.grupo_gmm.nunique()
k_gmm
```

```python
plt.figure(figsize=(18,8))

ax1 = plt.subplot(1, 2, 1)
sns.scatterplot(
    x="tsne_pca50_one", y="tsne_pca50_two",
    hue="grupo_kmeans",
    palette=sns.color_palette("hls", k_kmeans),
    data=df_kmeans,
    ax=ax1
)

ax2 = plt.subplot(1, 2, 2)
sns.scatterplot(
    x="tsne_pca50_one", y="tsne_pca50_two",
    hue="grupo_gmm",
    palette=sns.color_palette("hls", k_gmm),
    data=df_gmm,
    ax=ax2
)
ax1.title.set_text('KMeans')
ax2.title.set_text('GMM')
ax1.get_legend().remove()
ax2.get_legend().remove()

plt.show()
```
